```python
import random
from pathlib import Path
from spacy.training import Example
```

```python
TAG_MAP = {
'KN': {'tag': 'NOUN'},
'KK': {'tag': 'VERB'},
'ADJ': {'tag': 'ADJ'},
'KSN': {'tag': 'ADP'},
'KB': {'tag': 'AUX'},
'KG': {'tag': 'PRON'},
'KH': {'tag': 'CONJ'},
'ADV': {'tag': 'ADV'},
'KT': {'tag': 'question'},
'KBIL': {'tag': 'NUM'},
'KPM': {'tag': 'narrator'},
'KP': {'tag': 'command'},
'KAR': {'tag': 'direction'},
'PW': {'tag': 'discourse mark'},
'PP': {'tag': 'malay phrase'},
'NM': {'tag': 'non-malay word'},
'KEP': {'tag': 'short form'},
'UN': {'tag': 'unknown'},
'KNF': {'tag': 'deny'},
}
```

```python
import pandas as pd
import os
from tqdm import tqdm
import spacy
from spacy.tokens import Doc,DocBin

nlp = spacy.blank("xx") #load a new spacy model 
tagger = nlp.add_pipe("tagger") 

for tag, values in TAG_MAP.items():
    tagger.add_label(tag)
```

Train_data = []
for i in range(len(df1["word"])):
    Train_data.append((df1["word"][i],{"tags": list(df1["tags"][i])}))

```python
from spacy.lang.id import Indonesian
nlp.add_pipe("lemmatizer", lang= 'id',config={"mode": "lookup"})
nlp.initialize()
```

```python
l1=[]
l2=[]
for i in range(len(df1['word'])):
    l1.append(df1['word'][i])
    l2.append({"tags":[(df1['tags'][i])]})
    
Train_data=list(zip(l1,l2))
```

```python
print(Train_data)
```

```python
optimizer = nlp.begin_training()
#optimizer = nlp.initialize()
#optimizer = nlp.create_optimizer()
for i in range(25):
    random.shuffle(Train_data)
    losses = {}
    for text, annotations in Train_data:    
        example = Example.from_dict(nlp.make_doc(text), annotations)
        nlp.update([example], sgd=optimizer,losses=losses)
        print(losses)
    nlp.to_disk("C:\\Users\\User\\Desktop\\kelvin\\model")
```
